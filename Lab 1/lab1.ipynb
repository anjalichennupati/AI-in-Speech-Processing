```python
y, sr = librosa.load(audio_file)
librosa.display.waveshow(y, color='b')  # Specify color directly

plt.show()
```

```python
# A2. Observe the length & magnitude range of the signal from the plot. 
# Observing the plot, relate to the spoken words and silence between the words.
```

```python
duration = len(y) / sr  # Duration in seconds
print(f"Duration of the audio: {duration:.2f} seconds")
print(f"Magnitude range: {y.min()} to {y.max()}")
```
